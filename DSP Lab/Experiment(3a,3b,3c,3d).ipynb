# 3a line chart


```python
import matplotlib.pyplot as plt
import pandas as pd


height=[121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]
df=pd.DataFrame({"height":height,"weight":weight})
plt.xlabel('weight in kg')
plt.ylabel('height in cm')
plt.title('Average weight with respect to average height')
plt.plot(df.weight,df.height,marker='*',markersize=10,color='green',linewidth=2,linestyle='dashed')
plt.show()
```

#  3B(bar graph,histogram,area plot)


```python
#bar graph
import matplotlib.pyplot as plt
import pandas as pd


df=pd.read_csv("melasales.csv")
df.plot(kind='bar')
plt.title('Mela sales report')
plt.xlabel('Days')
plt.ylabel('Sales in rs')

plt.show()
```

```python
#HISTOGRAM
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,10,1,5,20])
df=pd.DataFrame(x)
df.plot(kind='hist',edgecolor="black")
plt.show()

```

```python
#AREA PLOT

import pandas as pd
import matplotlib.pyplot as plt
categories=['A','B','C','D']
values1=[5,7,3,4]
values2=[2,3,4,5]
plt.fill_between(categories,values1,color='skyblue',alpha=0.5,label='group1')
plt.fill_between(categories,values2,color='green',alpha=0.5,label='group2')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Area plot')
plt.legend()
plt.show()
```

# 3C(PIE CHARTS , BOX PLOT , SCATTER PLOT , BUBBLE PLOT)

```python
#PIE CHART
import pandas as pd
import matplotlib.pyplot as plt


df=pd.DataFrame({'GeoArea':[83743,78438,22327,22429,21081,16579,10486],'ForestCover':[67353,27692,17280,17321,19240,13464,8073]},
index=['Arunachal Pradesh','Assam','Manipur','Meghalaya',
'Mizoram','Nagaland','Tripura'])
df.plot(kind='pie',y='ForestCover',title='Forest cover of North Easternstates',legend=True)

plt.show()
```

```python
#BOX PLOT
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


data=pd.read_csv('marks.csv')
df=pd.DataFrame(data)
df.plot(kind='box')
plt.title('Performance Analysis')
plt.xlabel('Subjects')
plt.ylabel('Marks')


plt.show()
```

```python
#SCATTER PLOT
year=[2016,2017,2018,2019,2020,2021,2022]
population_in_cr=[6,7,8,9,10,11,12]
plt.scatter(year,population_in_cr,color='green',marker='*',linewidth=5)
```

```python
#BUBBLE PLOT
import pandas as pd
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[10,5,15,3,25]
plt.scatter(x,y,sizes=[50,20,100,30,10],color=["red","green","blue","yellow","brown"])
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from pywaffle import Waffle 
data ={'phone': ['Xiaomi', 'Samsung', 'Apple', 'Nokia'], 
       'stock': [20, 12, 8, 4] 
     }
df = pd.DataFrame(data)   
fig = plt.figure( 
    FigureClass = Waffle, 
    rows = 4, 
    values = df.stock, 
    labels = list(df.phone) 
)
```
