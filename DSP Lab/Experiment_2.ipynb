### 6. Aggregate functions

```python
df = pd.read_csv('class-grades.csv')
choices = np.array(['CSE','ECE','EEE'])
branches = []
for i in range(df.shape[0]):
    val = np.random.choice(choices)
    branches.append(val)
df['BRANCH'] = branches
```

```python
list = df.columns.tolist()
temp = list[-1]
list[-1] = list[0]
list[0] = temp
list
df = df.reindex(list,axis='columns')
```

```python
df.groupby(by=['BRANCH']).mean()
```

```python
df.iloc[10,5] = 55
df.groupby(by=['BRANCH']).count()
```

```python
df.groupby(['BRANCH']).sum()
```

```python
df.groupby(['BRANCH']).min()
```

```python
df.groupby(['BRANCH']).max()
```

```python
df.groupby(['BRANCH']).median()
```

```python
df.drop('BRANCH',axis=1,inplace=True)
```
