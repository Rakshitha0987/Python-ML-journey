```python
#PROBLEM-2
#binning
#q2
#read the income of a person divide into labels
import pandas as pd
n=int(input("enter the no. of peoples income do u want to enter:"))
l=[]
for i in range(n):
    income=int(input(f"enter the income {i+1}:"))
    l.append(income)
print(l)
bins=[0,10000,25000,50000,100000]
bin_names=['Basic Salary','Minimum Salary','Required Salary','Max Salary']
b=pd.cut(l,bins,labels=bin_names)
b
```

```python
#PROBLEM-3
#q3
#read the ages of a person and divide into labels
import pandas as pd
n=int(input("enter the no. of peoples age do u want to enter:"))
l1=[]
for i in range(n):
    age=int(input(f"enter the age of person {i+1}:"))
    l1.append(age)
print(l1)
bins=[0,13,19,27,60,100]
bin_names=['Child age','Teenage','Adult age','middle age','old age']
b=pd.cut(l1,bins,labels=bin_names)
b
```

```python
import pandas as pd

# Create a dictionary with student IDs and their marks
data = {
    'Student_ID': ['R200160', 'R200436', 'R201100', 'R200920', 'R200164'],
    'AT1': [8, 9, 7, 6, 8],
    'AT2': [7, 8, 6, 9, 7],
    'AT3': [9, 7, 8, 6, 9],
    'AT4': [6, 9, 7, 8, 6],
    'MID1': [15, 13, 11, 10, 12],
    'MID2': [14, 10, 12, 13, 11],
    'MID3': [13, 12, 10, 11, 13]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set 'Student_ID' as index
df.set_index('Student_ID', inplace=True)

# Operations
print("Max:")
print(df.max())
print("\nMin:")
print(df.min())
print("\nSum:")
print(df.sum())
print("\nMean:")
print(df.mean())
print("\nMedian:")
print(df.median())
print("\nMode:")
print(df.mode().iloc[0])  # mode() returns DataFrame, so we need to access the first row
print("\nStandard Deviation:")
print(df.std())
```
