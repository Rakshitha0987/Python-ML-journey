```python
import pandas as pd
data = {
    'Student_ID': ['R200260', 'R2000247', 'R200057', 'R200058', 'R200876'],
    'AT1': [8, 9, 7, 6, 8],
    'AT2': [7, 8, 6, 9, 7],
    'AT3': [9, 7, 8, 6, 9],
    'AT4': [6, 9, 7, 8, 6],
    'MID1': [15, 13, 11, 10, 12],
    'MID2': [14, 10, 12, 13, 11],
    'MID3': [13, 12, 10, 11, 13]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set 'Student_ID' as index
df.set_index('Student_ID', inplace=True)

# Operations
print("Max:")
print(df.max())
print("\nMin:")
print(df.min())
print("\nSum:")
print(df.sum())
print("\nMean:")
print(df.mean())
print("\nMedian:")
print(df.median())
print("\nMode:")
print(df.mode().iloc[0])  # mode() returns DataFrame, so we need to access the first row
print("\nStandard Deviation:")
print(df.std())
```

```python
#creating dataframe 
import numpy as np
a=np.arange(1,10,1)
b=np.arange(0,9,1)
name=pd.Series(["Pavan","Nithin","Prasad","Srujan","Arman","Govardhan","Venky","Anirudh","Tharun"])
age=pd.Series([20,18,19,19,18,20,19,18,20])
gender=pd.Series(["M","M","M","M","M","M","M","M","M"])
dept=pd.Series(["CSE","EEE","MECH","CSE","CSE","CSE","EEE","MECH","CIVIL"])
dept_df=pd.DataFrame({"NAME":name,"AGE":age,"GENDER":gender,"DEPT":dept})
dept_df
```

```python
#grouping based on department and then printing the first of all groups
gp1=dept_df.groupby("DEPT")
gp1.first()
```

```python
#size of the above group
gp1.size()
```

```python
#getting group
gp2=gp1.get_group("CSE")
gp2
```
