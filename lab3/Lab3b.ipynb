```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#AREA PLOT
category=['A','B','C','D']
value1=[5,3,7,2]
value2=[4,6,2,8]
plt.fill_between(category,value1,color="violet",alpha=1.0,label='Group1')
plt.fill_between(category,value2,color="green",alpha=0.5,label="Group2")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Area plot")

plt.show()
```

```python
#HISTOGRAM que1
x=np.array([1,2,3,4,10,1,15,20])
df3=pd.DataFrame(x)
df3.plot(kind="hist",edgecolor="black",color="red")
plt.show()
```

```python
#HISTOGRAM que2
data=[1,2,2,3,4,5,4,5,6,7,8,5,2,9,9,7,7,7,0,0]
plt.hist(data,bins=10,color="lightblue",edgecolor="black")
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Count")
plt.show()
```
