```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#LINECHART que1
height=[121.9,124.5,129.5,134.6,139.7,147.3,152.4,157.5,162.6]
weight=[19.7,21.3,23.5,25.9,28.5,32.1,35.7,39.6,43.2]
plt.plot(weight,height,color="Green",marker="*",markersize=10,linestyle="dashed",linewidth=2)
plt.xlabel("Weight in kg")
plt.ylabel("Height in cm")
plt.title("Average weight with respect to average height.")
plt.show()
```

```python
#LINE CHART que2
df=pd.DataFrame({"WEEK1":[5000,5900,6500,3500,4000,5300,7900],"WEEK2":[4000,3000,5000,5500,3000,4300,5900],"WEEK3":[4000,5800,3500,2500,3000,5300,6000]})
df.plot(kind="line",color=["Red","Blue","Brown"],marker="*")
plt.title("Mela Sales Report")
plt.xlabel("Days")
plt.ylabel("Sales in Rs")
plt.show()
```
