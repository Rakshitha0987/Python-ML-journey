```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#PIE CHART que1
x=np.array([360,361,25,120,80])
labels=["CSE","ECE","MECHANICAL","CIVIL","EEE"]
plt.pie(x,labels=labels)
plt.show()
```

```python
#PIE CHART que2
index=["Arunachal Pradesh","Assam","Manipur","Meghalaya","Mizoram","Nagaland","Tripura"]
df5=pd.DataFrame({"State":index,"GeoArea":[83743,78438,22327,22429,21081,16579,10486],"ForestCover":[67353,27692,17280,17321,19240,13464,8073]})
df5.plot(kind="pie",y="ForestCover",labels=index)
plt.title("Forest Cover of North Eastern States")
plt.show()
```

```python
#BOX PLOT
df4=pd.read_csv(r"/home/rguktrkvalley/Documents/Marks.csv")
df4.plot(kind="box",color="red")
plt.title("subject wise marks")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.show()
```

```python
#SCATTER PLOT que1
year=[2016,2017,2018,2019,2020,2021,2022,2023]
popu_in_cr=[5,6,7,8,9,10,11,12]
plt.scatter(year,popu_in_cr,color="r")
plt.show()
```

```python
#SCATTER PLOT que2
sale_in_rs=[5000,4000,2700,6400,9000,7300,4850]
discount=[10,5,9,6,15,7,8]
plt.scatter(sale_in_rs,discount,color="purple",marker="^")
plt.title("Discount of sales")
plt.xlabel("Sales")
plt.ylabel("Discount")
plt.show()
```

```python
#BUBBLE PLOT
x=[1,2,3,4,5]
y=[10,20,30,40,50]
plt.scatter(x,y,sizes=[10,20,30,40,50],color="black")
plt.show()
```
