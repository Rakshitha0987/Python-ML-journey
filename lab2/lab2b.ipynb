```python
import pandas as pd
import numpy as np
```

```python
id=pd.Series(["R200351","R200340","R200341","R200303","R200459","R200333"])
name=pd.Series(["RAKSHI","sri","pandu","abhi","gayathri","meghana"])
Class=[]
for i in range(0,6,1):
    Class.append("e2")
ph_no=pd.Series([8520992072,np.nan,8008862282,630541789,np.nan,np.nan])
address=pd.Series([np.nan,"Ananthapur",np.nan,np.nan,"Kurnool","Nellore"])
df=pd.DataFrame({"ID":id,"NAME":name,"CLASS":Class,"PHONE NO":ph_no,"PLACE":address})
df
```

```python
#checking if the columns have null values or not
df.isnull().any()
```

```python
df.info()
```

```python
#checking if each and every field is null
df.isnull()
```

```python
#checking if the columns have null values using column name
print(df["ID"].isnull().any())
print(df["PLACE"].isnull().any())
```

```python
#removing the rows that have null values
df.dropna()
```

```python
#removing the columns that have null values
df.dropna(axis=1)
```

```python
#filling the null values with 0
df.fillna("0")
```

```python
#filling the null values with earlier row values
df.ffill()
```

```python
#inputing missing values using median in phone no and a string instead of a place 
a=df["PHONE NO"].median()
df.fillna({"PHONE NO":a,"PLACE":"DMM"})
```
